```python
from selenium import webdriver
```

```python
#connect to web driver
driver = webdriver.Chrome('chromedriver.exe') 
```

```python
#opening the amazon website
url = 'https://www.amazon.in/'
driver.get(url)
```

```python
#Searching headphones in the search bar and clicking the search button
search_bar=driver.find_element_by_xpath('//input[@class ="nav-input nav-progressive-attribute"]')
search_bar.send_keys("headphones")
```

```python
#clicking the search buttin
driver.find_element_by_id('nav-search-submit-button').click()
```

```python
#creatimg empty lists
Product_URL = []
Ratings = []
Review = []
```

```python
for i in range(0,100):
    url=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a")
    for i in url:
        Product_URL.append(i.get_attribute('href'))
Product_URL
```

```python
for i in Product_URL:
    driver.get(i)
    
    #Clicking to see all reviews
    try:
        driver.find_element_by_xpath("//a[@class='a-link-emphasis a-text-bold']").click()
    except NoSuchElementException:
        pass    
    
    #Scrapping the details
    Start_page=1
    End_page=50
    for page in range(Start_page,End_page+1):

        try:
            Reviews=driver.find_elements_by_xpath("//span[@class='a-size-base review-text review-text-content']")
            for r in Reviews:
                Review.append(r.text.replace('\n',''))
        except NoSuchElementException:
            Review.append("Not Available")

        Rating=driver.find_elements_by_xpath("//div[@class='a-row']/a[1]")
        for i in Rating:
            rating=i.get_attribute('title')
            Ratings.append(rating) 
    

```

```python
#Checking the length of data extracted
print(len(Review),len(Ratings))
```

```python
#Saving in dataframe
headphones = pd.DataFrame({'Product_Review':Review[:50000],'Ratings':Ratings[:50000]})
```

```python
#Saving the data into a csv file
headphones.to_csv('Ratings_Data.csv')
```
